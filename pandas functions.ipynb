```python
import nbformat as nbf

# Create a new notebook
nb = nbf.v4.new_notebook()

# Add a title cell
nb.cells.append(nbf.v4.new_markdown_cell("# Python List Methods Examples"))

# Example 1: append()
append_code = '''
# append() - Adds an element to the end of the list
my_list = [1, 2, 3]
my_list.append(4)  # Append 4 to the list
my_list
'''
nb.cells.append(nbf.v4.new_code_cell(append_code))

# Example 2: extend()
extend_code = '''
# extend() - Extends the list by appending elements from another iterable
my_list = [1, 2, 3]
my_list.extend([4, 5, 6])  # Extend list by adding elements from another list
my_list
'''
nb.cells.append(nbf.v4.new_code_cell(extend_code))

# Example 3: insert()
insert_code = '''
# insert() - Inserts an element at a specific position
my_list = [1, 2, 3]
my_list.insert(1, 'a')  # Insert 'a' at index 1
my_list
'''
nb.cells.append(nbf.v4.new_code_cell(insert_code))

# Example 4: remove()
remove_code = '''
# remove() - Removes the first occurrence of a value
my_list = [1, 2, 3, 2]
my_list.remove(2)  # Remove the first occurrence of 2
my_list
'''
nb.cells.append(nbf.v4.new_code_cell(remove_code))

# Example 5: pop()
pop_code = '''
# pop() - Removes and returns an element at a specific position (default is the last item)
my_list = [1, 2, 3]
popped_element = my_list.pop()  # Remove and return the last element
my_list, popped_element
'''
nb.cells.append(nbf.v4.new_code_cell(pop_code))

# Example 6: clear()
clear_code = '''
# clear() - Removes all elements from the list
my_list = [1, 2, 3]
my_list.clear()  # Remove all elements from the list
my_list
'''
nb.cells.append(nbf.v4.new_code_cell(clear_code))

# Example 7: index()
index_code = '''
# index() - Returns the index of the first occurrence of a value
my_list = [1, 2, 3, 2]
index_of_value = my_list.index(2)  # Find the index of the first occurrence of 2
index_of_value
'''
nb.cells.append(nbf.v4.new_code_cell(index_code))

# Example 8: count()
count_code = '''
# count() - Returns the number of occurrences of a value
my_list = [1, 2, 3, 2]
count_of_value = my_list.count(2)  # Count the number of occurrences of 2
count_of_value
'''
nb.cells.append(nbf.v4.new_code_cell(count_code))

# Example 9: sort()
sort_code = '''
# sort() - Sorts the elements of the list in ascending order (in-place)
my_list = [3, 1, 2]
my_list.sort()  # Sort the list in ascending order
my_list
'''
nb.cells.append(nbf.v4.new_code_cell(sort_code))

# Example 10: reverse()
reverse_code = '''
# reverse() - Reverses the elements of the list (in-place)
my_list = [1, 2, 3]
my_list.reverse()  # Reverse the order of elements in the list
my_list
'''
nb.cells.append(nbf.v4.new_code_cell(reverse_code))

# Save the notebook to a file
with open('list_methods_examples.ipynb', 'w') as f:
    nbf.write(nb, f)
```
